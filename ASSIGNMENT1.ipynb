**Q6**) Calculate the Expected number of candies for a randomly selected child 
Below are the probabilities of count of candies for children (ignoring the nature of the child-Generalized view)
CHILD	Candies count	Probability

A  1 	 0.015

B	 4	 0.20

C	 3 	 0.65

D	 5 	 0.005

E	 6	 0.01

F	 2	 0.120

Child A – probability of having 1 candy = 0.015.
Child B – probability of having 4 candies = 0.20


```python
import pandas as pd
import numpy as np
```

```python
candies=(1,4,3,5,6,2)
probablity = (0.015,0.20,0.65,0.005,0.01,0.120)
df = pd.DataFrame(candies,columns=(['CANDIESCOUNT']))
df['PROBABILITY']=probablity
df['COUNT * PROBABILITY'] = df.CANDIESCOUNT * df.PROBABILITY
print('\n P(EXPECTED NUMBER OF CANDIES FOR A RANDOMLY PICKED CHILD) IS',np.sum(df['COUNT * PROBABILITY']))
```

**Q11)  Suppose we want to estimate the average weight of an adult male in    Mexico. We draw a random sample of 2,000 men from a population of 3,000,000 men and weigh them. We find that the average person in our sample weighs 200 pounds, and the standard deviation of the sample is 30 pounds. Calculate 94%,98%,96% confidence interval?**

```python
import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import norm
```

```python
print('Average weight of adult in mexico with 94% CI',stats.norm.interval(0.94,200,30/(2000**0.5)))
print('Average weight of adult in mexico with 98% CI',stats.norm.interval(0.98,200,30/(2000**0.5)))
print('Average weight of adult in mexico with 96% CI',stats.norm.interval(0.96,200,30/(2000**0.5)))
```
